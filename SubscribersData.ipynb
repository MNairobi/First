```python
# Select only numeric columns for calculating the mean
numeric_data = data.select_dtypes(include=['number'])
mean = numeric_data.mean()
print(mean)
```

```python
# Fill NaN values in the original DataFrame using the calculated means
data.fillna(mean, inplace=True)
data.isnull().sum()
```

```python
#fill  null valuea dor the non numerical colmns
non_numerical_cols = data.select_dtypes(exclude=['number']).columns
data[non_numerical_cols] = data[non_numerical_cols].fillna(data[non_numerical_cols].mode().iloc[0])

data.isnull().sum()
# categorical_cols = data.select_dtypes(include=['object']).columns

# # Fill NaN values with mode for each categorical column
# for col in categorical_cols:
#     data[col].fillna(data[col].mode()[0], inplace=True)

# data.isnull().sum()
```

```python
#Determine any outliers in your dataset either through visualization or using the numpy
#Z-score
import numpy as np
from scipy import stats
# Select only numeric columns for calculating the z-score
numeric_data = data.select_dtypes(include=['number'])

z_scores = np.abs(stats.zscore(numeric_data))
outliers = np.where(z_scores > 3)
print(outliers)
data.shape
```

```python
#Drop the rows with outliers
# data = data[(z_scores < 3).all(axis=1)]
# data.shape

import numpy as np
from scipy import stats

# Select only numeric columns for calculating the z-score
numeric_data = data.select_dtypes(include=['number'])

z_scores = np.abs(stats.zscore(numeric_data))
outlier_rows = np.where(z_scores > 3)[0]

# Drop the rows with outliers
data = data.drop(outlier_rows)
data.shape
```

```python
#Check for duplicates using an ideal method/function and remove the duplicates if any.
data.duplicated().sum()
```

```python
#remove duplicates
data = data.drop_duplicates()
data.duplicated().sum()
```
