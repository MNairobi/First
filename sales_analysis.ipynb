```python
###Let clean the data
```

```python
# Check for NaN values in the entire DataFrame
df_nul = df[df.isna().any(axis=1)]
# df_nul.head()

###let drop the null
df=df.dropna(axis=0,how='all')
df.head()
```

###the best month fors sales.how much was earned that month



Find the duplicates


```python
df=df[df['Order Date'].str[0:2]!='Or']
df.tail(5)
```

```python
df['month']=df['Order Date'].str[0:2]
df['month']=df['month'].astype('int32')
df.head()
```

let convert the columns into the right type

```python
df['Quantity Ordered']=pd.to_numeric(df['Quantity Ordered']) #make int
df['Price Each']=pd.to_numeric(df['Price Each']) #make float
df.head()
```

let get back to the question of the best month for sales


let create the sales column

```python
df['sales']=df['Quantity Ordered']*df['Price Each']
df.head()
```
